```
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```
dataset=pd.read_csv('/content/drive/My Drive/COVID-19 Dataset.csv')
```

```
dataset.head()
```

```
#Checking for null values in all columns
dataset.isnull().sum()
```

```
#REMOVING null values from "Active Cases"
dataset = dataset.dropna(axis=0, subset=['Active Cases'])
```

```
dataset.isnull().sum()
```

```
dataset = dataset[dataset.Date != "21-Apr-20"]
```

```
import plotly.express as px
px.scatter(dataset, x="Deaths", y="Recovered", animation_frame="Date", animation_group="Country", labels={"Country":"Country"},
           size="Active Cases", hover_name="Country", color="Active Cases",
           log_x=True, size_max=70, range_x=[1,100000], range_y=[1,100000])
```
